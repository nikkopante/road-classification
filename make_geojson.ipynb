```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("roadclassification.csv", )
df
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326"
)
gdf
```

```python
gdf.to_file("road_use_classification.geojson", driver="GeoJSON")  
```
