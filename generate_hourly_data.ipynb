```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Initialize start and end times
start_time = datetime.strptime("05/06/2023 05:00", "%m/%d/%Y %H:%M")
end_time = datetime.strptime("05/06/2023 23:00", "%m/%d/%Y %H:%M")

# Generate hourly intervals
time_intervals = []
current_time = start_time
while current_time < end_time:
    next_time = current_time + timedelta(hours=1)
    time_intervals.append({"startdate": current_time, "enddate": next_time})
    current_time = next_time

# Create DataFrame
df = pd.DataFrame(time_intervals)

# Display the DataFrame
df['startdate'] = df['startdate'].map(lambda x: x.isoformat())
df['enddate'] = df['enddate'].map(lambda x: x.isoformat())
df
```

```python
df.to_json('hourly_intervals.json', orient='records', indent=4)
```
